2. **Conditional Logic with Arrays:**
   Given the following arrays:
   ```python
   temperatures = np.array([25, 30, 15, 10, 20])
   ```
   - a. Create a new array `hot_or_cold` where values above 20 are labeled as "hot" and below 20 as "cold".
   - b. Create a new array where temperatures less than the mean of the entire array are replaced with -1.
   - c. Replace all temperatures below 15 with 0.
   - d. Given the following array `temperatures_2d = np.array([[25, 30, 15], [10, 20, 18], [22, 25, 21]])`. Replace values in each row with the maximum value of that row.
   - e. How long did it take you to solve this question? Was it easy, medium, or hard for you? Feel free to share any comments you have.
   
**Solution**
Certainly! Let's tackle each part of the problem:

a. **Create a new array `hot_or_cold`:**
To create the `hot_or_cold` array where values above 20 are labeled as "hot" and below 20 as "cold", you can use boolean indexing combined with np.where(). Here's how you can do it:

```python
import numpy as np

temperatures = np.array([25, 30, 15, 10, 20])

hot_or_cold = np.where(temperatures > 20, "hot", "cold")
print("Hot or Cold Array:", hot_or_cold)
```

This will output:
```
Hot or Cold Array: ['hot' 'hot' 'cold' 'cold' 'cold']
```

b. **Create a new array with temperatures less than the mean replaced with -1:**
To create a new array where temperatures less than the mean of the entire array are replaced with -1, you can use boolean indexing along with the mean function. Here's how you can do it:

```python
mean_temp = np.mean(temperatures)
new_array = np.where(temperatures < mean_temp, -1, temperatures)
print("New array with temperatures less than mean replaced with -1:", new_array)
```

c. **Replace temperatures below 15 with 0:**
To replace all temperatures below 15 with 0, you can use boolean indexing. Here's how you can do it:

```python
temperatures[temperatures < 15] = 0
print("Temperatures after replacing values below 15 with 0:", temperatures)
```

To replace values in each row of `temperatures_2d` with the maximum value of that row, you can use the np.max() function with the appropriate axis parameter. Here's how you can do it:

```python
temperatures_2d = np.array([[25, 30, 15], [10, 20, 18], [22, 25, 21]])

max_values_per_row = np.max(temperatures_2d, axis=1)
print("Maximum values in each row:", max_values_per_row)
```

Let's address each part of the problem:

a. **Compute the dot product of `matrix1` and `matrix2`:**
To compute the dot product of two matrices, you can use the `np.dot()` function. Here's how you can do it:

```python
import numpy as np

matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

dot_product = np.dot(matrix1, matrix2)
print("Dot product of matrix1 and matrix2:\n", dot_product)
```

b. **Find the determinant of `matrix1`:**
To find the determinant of a matrix, you can use the `np.linalg.det()` function. Here's how you can do it:

```python
determinant_matrix1 = np.linalg.det(matrix1)
print("Determinant of matrix1:", determinant_matrix1)
```

c. **Find the inverse of the sum of matrices `A` and `B`:**
To find the inverse of a matrix, you can use the `np.linalg.inv()` function. Here's how you can do it:

```python
sum_matrices = matrix1 + matrix2
inverse_sum_matrices = np.linalg.inv(sum_matrices)
print("Inverse of the sum of matrices A and B:\n", inverse_sum_matrices)
```

d. **Reshape `matrix1` into a 1D array:**
To reshape `matrix1` into a 1D array, you can use the `np.ravel()` function. Here's how you can do it:

```python
reshaped_matrix1 = np.ravel(matrix1)
print("Reshaped matrix1 into a 1D array:", reshaped_matrix1)
```

e. **Compute the eigenvalues and eigenvectors of the `matrix1`:**
To compute the eigenvalues and eigenvectors of a matrix, you can use the `np.linalg.eig()` function. Here's how you can do it:

```python
eigenvalues, eigenvectors = np.linalg.eig(matrix1)
print("Eigenvalues of matrix1:", eigenvalues)
print("Eigenvectors of matrix1:\n", eigenvectors)
```

f. **Calculate the cross-product of the `matrix1` and `vector=np.array([1, 2])`:**
To calculate the cross-product of a matrix and a vector, you can use the `np.cross()` function. Here's how you can do it:

```python
vector = np.array([1, 2])
cross_product = np.cross(matrix1, vector)
print("Cross-product of matrix1 and vector:", cross_product)
```

   Simulate a random walk with 100 steps using numpy. Start at position 0, and at each step, move either 1 step to the right or 1 step to the left with equal probability.
   - a. Plot the trajectory of the random walk.
   - b. Find the maximum distance from the starting position reached during the walk.
   - c. Calculate and print the mean and standard deviation of the final positions in the random walks.
   - d. Experiment with different step distributions. Instead of equal-sized coin flips, use `rng.standard_normal` to generate normally distributed steps with a mean and standard deviation. (e.g, draws = 0.25 * rng.standard_normal((nwalks, nsteps)))
   - e. How long did it take you to solve this question? Was it easy, medium, or hard for you? Feel free to share any comments you have.


**Solution**

a. **Plot the trajectory of the random walk:**
To simulate the random walk and plot its trajectory, you can use numpy for generating random steps and matplotlib for visualization. Here's how you can do it:

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulate random walk
n_steps = 100
steps = np.random.choice([-1, 1], size=n_steps)
positions = np.cumsum(steps)

# Plot trajectory
plt.plot(positions)
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('Random Walk Trajectory')
plt.grid(True)
plt.show()
```

b. **Find the maximum distance from the starting position reached during the walk:**
You can find the maximum distance from the starting position by taking the maximum absolute value of the positions array. Here's how you can do it:

```python
max_distance = np.max(np.abs(positions))
print("Maximum distance from the starting position:", max_distance)
```

c. **Calculate and print the mean and standard deviation of the final positions in the random walks:**
To calculate the mean and standard deviation of the final positions, you can use the `np.mean()` and `np.std()` functions. Here's how you can do it:

```python
mean_final_position = np.mean(positions)
std_final_position = np.std(positions)
print("Mean of final positions:", mean_final_position)
print("Standard deviation of final positions:", std_final_position)
```

d. **Experiment with different step distributions using `rng.standard_normal`:**
To simulate the random walk with normally distributed steps, you can use `np.random.standard_normal()` to generate normally distributed steps. Here's how you can modify the code:

```python
# Simulate random walk with normally distributed steps
n_steps = 100
steps = 0.25 * np.random.standard_normal(size=n_steps)
positions = np.cumsum(steps)

# Plot trajectory
plt.plot(positions)
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('Random Walk Trajectory with Normally Distributed Steps')
plt.grid(True)
plt.show()
```

This code will generate a random walk where the steps are drawn from a normal distribution with mean 0 and standard deviation 0.25. You can adjust the mean and standard deviation as needed.

   Given the DataFrame `employee_data`:
   ```python
   employee_data = pd.DataFrame({"Department": ["IT", "HR", "IT", "Finance", "HR"], "Salary": [60000, 55000, 65000, 70000, 60000]})
   ```
   - a) Group the DataFrame `employee_data` by 'Department' and calculate the total salary for each department.
   - b) Find the department with the highest average salary.
   - c) Calculate the difference between the highest and lowest salary in each department.
   - d) How long did it take you to solve this question? Was it easy, medium, or hard for you? Feel free to share any comments you have.
   
**Solution**

a) **Group the DataFrame `employee_data` by 'Department' and calculate the total salary for each department:**
You can use the `groupby()` function to group the DataFrame by the 'Department' column and then use the `sum()` function to calculate the total salary for each department. Here's how you can do it:

```python
import pandas as pd

employee_data = pd.DataFrame({"Department": ["IT", "HR", "IT", "Finance", "HR"], "Salary": [60000, 55000, 65000, 70000, 60000]})

total_salary_by_department = employee_data.groupby('Department')['Salary'].sum()
print("Total salary for each department:\n", total_salary_by_department)
```

b) **Find the department with the highest average salary:**
You can use the `groupby()` function to group the DataFrame by the 'Department' column and then use the `mean()` function to calculate the average salary for each department. After that, you can use the `idxmax()` function to find the department with the highest average salary. Here's how you can do it:

```python
average_salary_by_department = employee_data.groupby('Department')['Salary'].mean()
department_highest_avg_salary = average_salary_by_department.idxmax()
print("Department with the highest average salary:", department_highest_avg_salary)
```

c) **Calculate the difference between the highest and lowest salary in each department:**
You can use the `groupby()` function to group the DataFrame by the 'Department' column and then use the `max()` and `min()` functions to calculate the highest and lowest salary for each department. After that, you can calculate the difference between them. Here's how you can do it:

```python
max_salary_by_department = employee_data.groupby('Department')['Salary'].max()
min_salary_by_department = employee_data.groupby('Department')['Salary'].min()

salary_difference_by_department = max_salary_by_department - min_salary_by_department
print("Difference between highest and lowest salary in each department:\n", salary_difference_by_department)
```
